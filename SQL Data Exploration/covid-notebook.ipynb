### clean the two tables CovidDeaths.csv and CovidVaccinations.csv before you import them into mysql tables

replace blank cells with 'NULL' except last column, replace blank cells with zero for last column.

Also, remember to reformat date field to be in mysql accepted format yyyy-mm-dd.

```python
import pandas as pd
```

#### clean CovidVaccinations.csv

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("F:\Library\Engineering\computer science\Data Science\portfolio projects\Project 4 - SQL Data Exploration - part 1\datasets\CovidVaccinations.csv")
```

```python
# Replace blank cells with NULL
df = df.map(lambda x: 'NULL' if pd.isna(x) or x == '' else x)
```

```python
# Save the updated DataFrame to a new CSV file
df.to_csv('datasets/CovidVaccination - Cleaned.csv', index=False)
```
